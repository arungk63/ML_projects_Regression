### Data cleaning 

```python
#Year
#Selecting only numeric values from yesr by using string method
cars = cars[cars['year'].str.isnumeric()]
```

```python
#Converting the yesr to int 

cars['year']=cars['year'].astype(int)
```

```python
#Price

cars=cars[cars['Price']!='Ask For Price']
```

```python
cars['Price'] =cars['Price'].str.replace(',','').astype(int)
```

```python
cars.info()
```

```python
# KM driven
cars['kms_driven']
```

```python
#removing the kms and ',' from the data
cars['kms_driven'] =cars['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
cars['kms_driven']
```

```python
cars['kms_driven']= cars['kms_driven'].astype(int)
```

```python
cars.info()
```

```python
#Fuel type
cars = cars[~cars['fuel_type'].isna()]
```

```python
cars['name'] =cars['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
cars.head()
```

```python
cars.reset_index(drop=True, inplace=True)
```

### ML Model - Linear Regression

```python
x= cars.drop(['Price'],axis=1)
y=cars['Price']
```

```python
#importing the ML model and preprocess packages
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
```

```python
#Test train split 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
ohe = OneHotEncoder()
```

```python
ohe.fit(x[['name','company','fuel_type']])
```

```python
#Data processing using cloumn tranformer and pipeline
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
col_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                              remainder='passthrough')
```

```python
lr =LinearRegression()
```

```python
pipe = make_pipeline(col_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred) 
```

```python
for i in range(10):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=i)
    lr =LinearRegression()
    pipe = make_pipeline(col_trans,lr)
    pipe.fit(x_train,y_train)
    y_pred = pipe.predict(x_test)
    print(f'The r2 score is {r2_score(y_test,y_pred)}, {i} ')
    
```

```python
#Getting the random state with max r2 score 
```

```python
score = []
for i in range(1000):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=i)
    lr =LinearRegression()
    pipe = make_pipeline(col_trans,lr)
    pipe.fit(x_train,y_train)
    y_pred = pipe.predict(x_test)
    score.append(r2_score(y_test,y_pred))
```
